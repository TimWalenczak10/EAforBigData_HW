```python
class Node:
    """
    Tree node with left and right child and data
    """
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
```

```python
class Tree:
    def __init__(self):
        """Create an empty binary search tree"""
        self.root = None
    
    def insert(self, item):
        """Add an item in the binary search tree"""
        if self.root is None:
            self.root = Node(item)
        else:
            root = self.root
            node = self.root
            while node is not None:
                if (node.data < item):
                    root = node
                    node = node.right
                else:
                    root = node
                    node = node.left
            if (root.data < item):
                root.right = Node(item)
            else:
                root.left = Node(item)
    
    def find(self, item):
        """Returns true if item is in the tree or false otherwise"""
        if (self.root == item):
            return True
        else:
            node = self.root
            while node is not None:
                if (node.data == item):
                    return True
                elif (node.data < item):
                    node = node.right
                else:
                    node = node.left
        return False
        
```

```python
tree = Tree()
```

```python
tree.insert(30)
tree.insert(5)
tree.insert(50)
tree.insert(8)
tree.insert(3)
tree.insert(100)
tree.insert(40)
```

```python
tree.find(8)
```
